```python
#Passo 2
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

```python
#Passo 3
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))

driver.get("https://www.etf.com/etfanalytics/etf-finder")
```

```python
#Passo 5.1
time.sleep(10)

botao_100 = driver.find_element("xpath",
                               '//*[@id="panel:r0:0"]/div/div[2]/div[1]/div/div[5]/button')

#botao_100.click()

driver.execute_script("arguments[0].click();", botao_100)
```

```python
#Passo 5.2
numero_paginas = driver.find_element("xpath", '//*[@id="panel:r0:0"]/div/div[2]/div[2]/ul/li[8]')

numero_paginas = int(numero_paginas.text.replace("of ", ""))
```

```python
print(numero_paginas)
```

```python
lista_tabela_por_pagina =[]

elemento = driver.find_element("xpath", '//*[@id="screener-output-table"]')
```

```python
#Passo 6.1
lista_tabela_por_pagina =[]

elemento = driver.find_element("xpath", '//*[@id="screener-output-table"]')

for i in range(1, numero_paginas + 1):
    
    html_tabela = elemento.get_attribute('outerHTML')
    
    tabela = pd.read_html(str(html_tabela))[0]
    
    lista_tabela_por_pagina.append(tabela)
    
    botao_avancar_pagina = driver.find_element("xpath", '//*[@id="panel:r0:0"]/div/div[2]/div[2]/ul/li[9]/a')
    
    driver.execute_script("arguments[0].click();", botao_avancar_pagina)
    
    
tabela_cadastro_etfs = pd.concat(lista_tabela_por_pagina)
```

```python
#Passo 6.2
#formulario_de_voltar_pagina = driver.find_element("xpath", '//*[@id="goToPage"]')

#formulario_de_voltar_pagina.clear()
#formulario_de_voltar_pagina.send_keys("1")
#formulario_de_voltar_pagina.send_keys(u'\ue007')

pagina_um = driver.find_element("xpath",
                               '//*[@id="panel:r0:0"]/div/div[2]/div[2]/ul/li[2]/a')

driver.execute_script("arguments[0].click();", pagina_um)
```

```python
#Passo 6.3
botao_mudar_para_performance = driver.find_element("xpath", '//*[@id="tab:r0:1"]')

botao_mudar_para_performance.click()

lista_tabela_por_pagina =[]

elemento = driver.find_element("xpath", '//*[@id="screener-output-table"]')

for i in range(1, numero_paginas + 1):
    
    html_tabela = elemento.get_attribute('outerHTML')
    
    tabela = pd.read_html(str(html_tabela))[0]
    
    lista_tabela_por_pagina.append(tabela)
    
    botao_avancar_pagina = driver.find_element("xpath", '//*[@id="panel:r0:1"]/div/div[2]/div[2]/ul/li[9]')
    
    #botao_avancar_pagina.click()
    
    driver.execute_script("arguments[0].click();", botao_avancar_pagina)
    

tabela_rentabilidade_etfs = pd.concat(lista_tabela_por_pagina)
```

```python
driver.quit()
```
